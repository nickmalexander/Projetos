**5. Considerando a bases de dados `base_pessoal.csv`, crie duas novas colunas. A primeira deve se chamar `primeiro_nome` e só conter os primeiros nomes dos clientes; a segunda coluna deve se chamar `sobrenome` e só conter os sobrenomes dos clientes.**

```python
primeiro = []
sobrenome = []
for i in range(df_pessoal.shape[0]):
  a = df_pessoal.nome.iloc[i]
  t = a.find(" ")
  primeiro.append(a[:t])
  sobrenome.append(a[t:])
df_pessoal["primeiro_nome"] = primeiro
df_pessoal["sobrenome"] = sobrenome
df_pessoal
```

**7. Considerando todas bases de dados, faça com que todas as variáveis

**10. Qual é a idade mediana de nossos clientes?**

```python
from datetime import date
import statistics

def calculateAge(birthDate):
    today = date.today()
    fgh= [int(x) for x in birthDate.split('-')]
    age = today.year - fgh[0] - ((today.month, today.day) < (fgh[1], fgh[2]))

    return age

datas_nascimento = [calculateAge(df_pessoal.nascimento.iloc[i]) for i in range(df_pessoal.shape[0])]
idades = sorted(datas_nascimento)
statistics.median(idades)
```
